```python
# Dependencies
import numpy as np
import pandas as pd
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
#import mats for graphs
%matplotlib inline
from matplotlib import style
#https://www.analyticsvidhya.com/blog/2021/08/exploring-matplotlib-stylesheets-for-data-visualization/
#https://matplotlib.org/stable/gallery/style_sheets/dark_background.html
import matplotlib.pyplot as plt
style.use('dark_background')
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)

# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

## D1: Determine the Summary Statistics for June

```python
# 1. Import the sqlalchemy extract function.
from sqlalchemy import extract

# 2. Write a query that filters the Measurement table to retrieve the temperatures for the month of June. 
june_temp = session.query(Measurement.date, Measurement.tobs).filter(extract('month', Measurement.date) == 6).all()

print(june_temp)

```

```python
#  3. Convert the June temperatures to a list.
june_temp_list = session.query(Measurement.tobs).filter(extract('month', Measurement.date) == 6).all()
print(june_temp_list)
```

```python
# 4. Create a DataFrame from the list of temperatures for the month of June. 
june_temp_df = pd.DataFrame(june_temp_list, columns=['June Temps'])
print(june_temp_df)
```

```python
# 5. Calculate and print out the summary statistics for the June temperature DataFrame.
june_temp_df.describe()
```

## D2: Determine the Summary Statistics for December

```python
# 6. Write a query that filters the Measurement table to retrieve the temperatures for the month of December.
dec_temp = session.query(Measurement.date, Measurement.tobs).filter(extract('month', Measurement.date) == 12).all()

print(dec_temp)
```

```python
# 7. Convert the December temperatures to a list.
dec_temp_list = session.query(Measurement.tobs).filter(extract('month', Measurement.date) == 12).all()
print(dec_temp_list)
```

```python
# 8. Create a DataFrame from the list of temperatures for the month of December. 
dec_temp_df= pd.DataFrame(dec_temp_list, columns=['December Temps'])
print(dec_temp_df)
```

```python
# 9. Calculate and print out the summary statistics for the Decemeber temperature DataFrame.
dec_temp_df.describe()
```

```python
#JUNE 
#creating visuals- to better explain the data 
#histogram used since its condences the ploted data- to an easy on the eyes graph
june_temp_df = pd.DataFrame(june_temp_list, columns=['tobs'])
june_temp_df.plot.hist(bins=15)
#10 bins- clunky 
#20 bins- too short
plt.tight_layout()
plt.title('June Temperatures')
plt.xlim([60, 90])
plt.ylim([0,350])
```

```python
#DF of precipitation in june- the 
june_temp_precip = session.query(Measurement.tobs,Measurement.prcp).filter(extract('month', Measurement.date) == 6).all()
june_temp_precip_df = pd.DataFrame(june_temp_precip, columns=['June Temps', 'Precipitation'])
june_temp_precip_df.describe()
```

```python
june_temp_precip_df = pd.DataFrame(june_temp_precip, columns=['tobs','precip'])
june_temp_prcp_df = june_temp_precip_df.dropna()
june_temp_precip_df.plot.scatter('tobs', 'precip')
plt.tight_layout()
plt.title('June Temperature v Precipitation')
plt.xlim([60, 85])
plt.ylim([-1,5])
```
